```python
import numpy as np
import matplotlib.pyplot as plt
from WPS_Models import phipp_models
import pandas as pd
```

```python
# Step 1: Load your data
BLdata = pd.read_csv('TA10_BLparams_zones.csv', header='infer', sep=' ')
```

```python
# Step 2: Create an instance of the model
model = phipp_models.PhippModels('rozenberg', BLdata.iloc[15],normalization='frequency',spectral_normalization=2e-5)

```

```python
# Step 3: Define a frequency array (e.g., 0 Hz to 10000 Hz with 500 points)
frequencies = np.linspace(0, 10000, 500)

# Step 4: Compute the PSD using your model
psd = model.compute_model(frequencies)
```

```python


# Step 5: Plotting
plt.figure(figsize=(8,5))
plt.semilogx(frequencies, psd)
plt.xlabel('Frequency [Hz]')
plt.ylabel('PSD [dB]')
plt.grid(True)
plt.xlim(100, 10000)
plt.ylim(0,100)
plt.show()
```
